```python
import cv2
import glob
```

```python
for filename in glob.glob("results/yes_object_results/*"):
    print(filename.split())
#     cv2.imshow('results', cv2.imread(filename))
#     cv2.waitKey(50)
# cv2.destroyAllWindows()
```

```python
nums = [filename.split(".")[0].split("\\")[-1] for filename in glob.glob("results/yes_object_results/*")]
names = glob.glob("results/yes_object_results/*")

fileDict = {}
for a,b in zip(names, nums):
    fileDict[int(b)] = a
```

```python
img_array = []
for a in sorted(fileDict):
    img_array.append(cv2.resize(cv2.imread(fileDict[a]), (1000,500), interpolation = cv2.INTER_AREA))
#     cv2.imshow('results', cv2.resize(cv2.imread(fileDict[a]), (1000,500), interpolation = cv2.INTER_AREA)  )
#     cv2.waitKey(100)
# cv2.destroyAllWindows()
    
```

```python

out = cv2.VideoWriter('project.avi',cv2.VideoWriter_fourcc(*'DIVX'), 15, (1000,500))
 
for i in range(len(img_array)):
    out.write(img_array[i])
out.release()
```
